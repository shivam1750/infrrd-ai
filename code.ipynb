### ` Data Analysis:`

#### 1. Loading Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import yfinance  as yf
import plotly.express as px
import plotly.graph_objs as go
import seaborn as sns
```

```python
stock_data = yf.Ticker("TSLA")
data = stock_data.history(period="1y")
data.head()
```

```python
data.shape
```

```python
data.info()
```

#### 2. Closing prices of the stock

```python
data['Daily Return'] = data['Close'].pct_change()
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna()
```

```python
data.describe()
```

```python
px.line(data, 
              x=data.index, 
              y='Close', 
              title='Tesla Stock Closing Price (Last 1 Year)', 
              labels={'Close': 'Price (USD)', 'index': 'Date'}, 
              line_shape='linear')
```

#### 3. Plotting Stock Data : 

```python
week_52_high = data['Close'].max()
week_52_low = data['Close'].min()
print(week_52_high,"\n",week_52_low)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

fig.add_trace(go.Scatter(
    x=[data['Close'].idxmax()], y=[week_52_high],
    mode='markers+text', marker=dict(color='green', size=12),
    text="52-Week High", textposition='top center'
))

fig.add_trace(go.Scatter(
    x=[data['Close'].idxmin()], y=[week_52_low],
    mode='markers+text', marker=dict(color='red', size=12),
    text="52-Week Low", textposition='bottom center'
))

fig.update_layout(
    title='Tesla Stock Closing Price (Last 1 Year)',
    xaxis_title='Date',
    yaxis_title='Price (USD)',
    showlegend=False
)
fig.show()
```

#### 5. Weekly High-Low Analysis:


```python
weekly_data = data.resample('W').agg({'High': 'max', 'Low': 'min'})
weekly_data.head()
```

```python
weekly_data['Absolute Difference'] = weekly_data['High'] - weekly_data['Low']
weekly_data['Percent Difference'] = (weekly_data['Absolute Difference'] / weekly_data['Low']) * 100
weekly_data.head()
```

```python
def plot_weekly_high_low(threshold: float, difference_type: str = "absolute"):
    if difference_type == "absolute":
        filtered_data = weekly_data[weekly_data['Absolute Difference'] > threshold]
    elif difference_type == "percent":
        filtered_data = weekly_data[weekly_data['Percent Difference'] > threshold]
    else:
        raise ValueError("Value Error")
    
    filtered_data = filtered_data.reset_index()

    fig = px.bar(
        filtered_data, 
        x='Date', 
        y=['High', 'Low'], 
        title=f"Weekly High & Low Prices (Filter by {difference_type.capitalize()} > {threshold})", 
        barmode='group', 
        labels={'value': 'Price (USD)', 'Date': 'Week'},
    )
    fig.show()
```

```python
plot_weekly_high_low(threshold=20, difference_type="absolute")
```

```python
plot_weekly_high_low(threshold=5, difference_type="percent")
```
